```python
import rdflib
g = rdflib.Graph()

g.parse('TMDB.ttl', format='turtle')

query_a = g.query(
"""
SELECT DISTINCT ?gg
WHERE {
    ?m a :Movie ;
    :genres ?g .

?g :name ?gg .
}
"""
)

for row in query_a:
    print(row[0])
```

```python
query_b = g.query(
"""
SELECT DISTINCT ?d
WHERE {
    ?c a :Crew ;
    :department ?d .
}
"""
)
for row in query_b:
    print(row[0])
```

```python
query_c = g.query(
"""
SELECT (COUNT(DISTINCT ?company) AS ?companyCount)
    WHERE {
        ?company a :Company .
    }
"""
)
for row in query_c:
    print(row[0])
```

```python
query_d = g.query(
"""
SELECT DISTINCT ?id
WHERE {
    ?cn a :Collection ;
    :id ?id .
}
"""
)
for row in query_d:
    print(row[0])
```

```python
query_e = g.query(
"""
SELECT DISTINCT ?title
WHERE {
    ?m a :Movie ;
    :title ?title ;
    :original_language ?lang .
    FILTER(?lang = "de")
}
"""
)
for row in query_e:
    print(row[0])
```

```python
query_f = g.query(
"""
SELECT DISTINCT ?title
WHERE {
    ?m a :Movie ;
    :title ?title ;
    :Keywords ?keyword .
    ?keyword :name ?name .
    FILTER(?name = "julius caesar")
}
"""
)
for row in query_f:
    print(row[0])
```

```python
query_g = g.query(
"""
SELECT DISTINCT ?name
WHERE {
    ?c a :Cast ;
    :name ?name .
    ?crew :job ?job .
    FILTER(?job = "Director")
}
"""
)
for row in query_g:
    print(row[0])
```

```python
query_h = g.query(
"""
SELECT ?character
WHERE {
    ?c a :Cast ;
    :character ?character ;
    :name ?name .
    FILTER(?name = "Kevin Spacey")
}
"""
)
for row in query_h:
    print(row[0])
```

```python
query_i_common = g.query(
"""
SELECT ?original_title ?id ?budget ?imdb_id ?original_language ?overview ?popularity ?poster_path ?release_date ?runtime ?status ?tagline ?title ?revenue
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;
    :id ?id ;
    :budget ?budget ;
    :imdb_id ?imdb_id ;
    :original_language ?original_language ;
    :overview ?overview ;
    :popularity ?popularity ;
    :poster_path ?poster_path ;
    :release_date ?release_date ;
    :runtime ?runtime ;
    :status ?status ;
    :tagline ?tagline ;
    :title ?title ;
    :revenue ?revenue ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)
query_i_genre = g.query(
"""
SELECT ?genre_Name
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;

    :genres ?genres .
    ?genres :name ?genre_Name ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)
query_i_production_companies = g.query(
"""
SELECT ?production_companies_Name
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;

    :production_companies ?production_companies .
    ?production_companies :name ?production_companies_Name ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)
query_i_production_countries = g.query(
"""
SELECT ?production_countries_Name
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;

    :production_countries ?production_countries .
    ?production_countries :name ?production_countries_Name ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)
query_i_spoken_languages = g.query(
"""
SELECT ?spoken_languages_Name
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;

    :spoken_languages ?spoken_languages .
    ?spoken_languages :name ?spoken_languages_Name ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)
query_i_keywords = g.query(
"""
SELECT ?production_countries_Name
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;


    :Keywords ?Keywords .
    ?Keywords :name ?Keywords_Name ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)
query_i_cast = g.query(
"""
SELECT ?production_countries_Name
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;

    :cast ?cast .
    ?cast :name ?cast_Name ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)
query_i_crew = g.query(
"""
SELECT ?production_countries_Name
WHERE {
    ?m a :Movie ;
    :original_title ?original_title ;

    :crew ?crew .
    ?crew :name ?crew_Name ;

    FILTER(?original_title = "Ruby in Paradise")
}
"""
)

for row in query_i_common:
    for i in range(0, len(row)):
        print(row[i])

for row in query_i_genre:
    for i in range(0, len(row)):
        print(row[i])
for row in query_i_spoken_languages:
    for i in range(0, len(row)):
        print(row[i])
for row in query_i_production_companies:
    for i in range(0, len(row)):
        print(row[i])
for row in query_i_production_countries:
    for i in range(0, len(row)):
        print(row[i])
for row in query_i_keywords:
    for i in range(0, len(row)):
        print(row[i])
for row in query_i_cast:
    for i in range(0, len(row)):
        print(row[i])
for row in query_i_crew:
    for i in range(0, len(row)):
        print(row[i])
```
